```python
import random
import tkinter as tk
from PIL import Image, ImageTk

# Function to determine winner
def determine_winner(user_choice):
    choices = ["Rock", "Paper", "Scissors"]
    computer_choice = random.randint(0, 2)
    computer_choice_label.config(image=choice_photos[computer_choice])
    if computer_choice == user_choice:
        return "It's a draw !!"
    elif computer_choice == 0 and user_choice == 2:
        return "You lose !!"
    elif user_choice == 0 and computer_choice == 2:
        return "You win !!!"
    elif computer_choice > user_choice:
        return "You lose !!"
    elif user_choice > computer_choice:
        return "You win !!!"

# Function to handle user choice
def handle_choice(choice):
    user_choice = int(choice)
    user_choice_label.config(image=choice_photos[user_choice])
    result_label.config(text=determine_winner(user_choice))

# Create GUI window
root = tk.Tk()
root.title("Rock-Paper-Scissors Game")

# Create frame for choices at the top center
choices_frame = tk.Frame(root)
choices_frame.pack(side=tk.TOP, pady=10)

# Create and pack labels for choices
instruction_label = tk.Label(choices_frame, text="Enter your choice:", font=("Arial", 16))
instruction_label.pack()

# Load images
rock_img = Image.open("rock.png")
paper_img = Image.open("paper.png")
scissors_img = Image.open("scissor.png")

# Convert images to Tkinter PhotoImage objects
rock_photo = ImageTk.PhotoImage(rock_img)
paper_photo = ImageTk.PhotoImage(paper_img)
scissors_photo = ImageTk.PhotoImage(scissors_img)

# Create buttons with images
rock_button = tk.Button(choices_frame, image=rock_photo, command=lambda: handle_choice(0))
rock_button.pack(side=tk.LEFT, padx=10)
paper_button = tk.Button(choices_frame, image=paper_photo, command=lambda: handle_choice(1))
paper_button.pack(side=tk.LEFT, padx=10)
scissors_button = tk.Button(choices_frame, image=scissors_photo, command=lambda: handle_choice(2))
scissors_button.pack(side=tk.LEFT, padx=10)

# Create frame for user and computer choices
choices_result_frame = tk.Frame(root)
choices_result_frame.pack(pady=20)

# Create labels for user's choice and computer's choice
user_choice_label = tk.Label(choices_result_frame, text="Your Choice:", font=("Arial", 14))
user_choice_label.grid(row=0, column=0, padx=10)

computer_choice_label = tk.Label(choices_result_frame, text="Computer's Choice:", font=("Arial", 14))
computer_choice_label.grid(row=0, column=1, padx=10)

# Create labels for displaying images
user_image_label = tk.Label(choices_result_frame, image=None)
user_image_label.grid(row=1, column=0, padx=10)

computer_image_label = tk.Label(choices_result_frame, image=None)
computer_image_label.grid(row=1, column=1, padx=10)

# Create result label
result_label = tk.Label(root, text="", font=("Arial", 16))
result_label.pack()

# Dictionary to map choice index to corresponding photo
choice_photos = {
    0: rock_photo,
    1: paper_photo,
    2: scissors_photo
}

# Run the GUI
root.mainloop()
```
